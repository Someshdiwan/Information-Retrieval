```python
Text Preprocessing Theory in Brief

1. Preprocessing Framework Overview

   - Tokenization: Splitting text into smaller units like words or sentences. Challenges include handling punctuation, abbreviations, contractions, etc.
   - Normalization: Standardizing text (e.g., lowercase conversion, removing numbers/punctuation, and stripping whitespace).
   - Substitution: Replacing or removing elements, such as stop words, sparse terms, or noise.

---

2. Key Steps in Text Preprocessing

1. Tokenization: Breaking text into tokens (e.g., words). 

- Example: "Dr. Ford ran." → ["Dr.", "Ford", "ran"]
   
2. Normalization: Preparing text for analysis.
   
      - Convert to lowercase: "HELLO" → "hello"

      - Remove punctuation: "I, me!" → "I me"

      - Stemming: Reduce words to root form (e.g., "running" → "run").

      - Lemmatization: Map words to their canonical form (e.g. "better" → "good").

3. Stop Word Removal: Exclude common words (e.g., "the", "and") that add little meaning.
   
4. Noise Removal: Eliminate irrelevant elements, e.g., special characters, task-specific noise.

---

3. Stemming vs Lemmatization

   - Stemming: Strips affixes; faster but less accurate.
     
   - "running" → "run"
   
   - Lemmatization: Uses linguistic rules; more accurate but slower.
     
   - "better" → "good"

---

Example usage

text = "I am running and eating. I like to play games, running and eating."

preprocessed_text = preprocess_text(text)

print(preprocessed_text)

---

5. Output Example

Input: "I am running and eating. I like to play games, running and eating."

Output: "run eat like play game run eat"

---

6. Why Preprocess Text?

- Simplifies text for machine learning or NLP tasks.
- Reduces dimensionality by focusing on meaningful components.
- Improves accuracy and efficiency in text-based models.
```

```python
import spacy

# Load the SpaCy model
nlp = spacy.load("en_core_web_sm")

def preprocess_text(text):
    doc = nlp(text)

    # Tokenize, remove stop words, and lowercase
    tokens = [token.lemma_.lower() for token in doc if not token.is_stop and token.is_alpha]

    # Join tokens back into a single string
    return " ".join(tokens)

text = "I am running and eating. I like to play games, running and eating."
preprocessed_text = preprocess_text(text)
print(preprocessed_text)
```
